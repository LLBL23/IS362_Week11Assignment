## Retrieving the Most Emailed Articles for the Last Day
To retreive the most emailed articles for the last day I made a request to the API, ensured the request was successful, saved the data as a list, and saved the list as a DataFrame. The same process was repeated to retrieve the data for the most shared and the most viewed articles.

```python
most_emailed_url = f"{Root_url}/viewed/1.json"

# request
response = requests.get(most_emailed_url, params=params, headers=headers)

# check status
if response.status_code == 200:
    data = response.json()
else:
    print(f"Error: {response.status_code}")
    print(response.text)
    
# save as a list
most_emailed_articles = data.get("results", [])

# save as DataFrame
df = pd.DataFrame(most_emailed_articles)

# print DataFrame
df
```
